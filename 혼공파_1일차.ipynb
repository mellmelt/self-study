## 식별자

식별자는 이름을 붙일 때 사용하는 단어이다. 주로 변수 또는 함수 이름 등으로 사용된다. 식별자는 다음 규칙을 지켜 만들어야 한다.

- 키워드를 사용하면 안됨
- 특수문자는 언더바(_)만 허용됨
- 숫자로 시작하면 안됨
- 공백을 포함할 수 없음

공백을 넣지 않고 식별자를 이해하기 힘든 경우가 있다. 이를 해결하기 위하여 두 가지 방법을 사용할 수 있다

1. 스네이크 케이스(snake_case) : 단어 사이에 언더바(_) 기호 붙이기 *ex) item_list*

2. 캐멀 케이스(CamelCase) : 단어의 첫 글자를 대문자로 만들기 *ex) ItemList. 'itemList'와 같은 경우는 파이썬에서 사용하지 않는다*
